```python
def city():
    print("Welcome to Book My Show Application!!")
    
    print("Choose Your City \n 1. Indore \n 2. Bhopal \n 3. Mumbai \n 4. Bangalore")
    c = int(input("Enter you city:"))
    if c == 1:
        print("Your city is Indore")
    elif c == 2:
        print("Your city is Bhopal")
    elif c == 3:
        print("Your City is Mumbai")
    elif c == 4:
        print("Your city is Bangalore")
    else:
        print("Select the valid option")
        city()
    
    t_movie()
```

```python
def t_movie():
    print("Choose the movie which you want to watch \n 1. Govinda \n 2. Ghost house \n 3. Love Marriage \n 4. back")
    t = int(input("Enter your movie name:"))
    if t == 1:
        print("You have choosen Govinda")
    elif t == 2:
        print("You have choosen Ghost House")
    elif t == 3:
        print("You have choosen Love Marriage")
    elif t == 4:
        city()
    else:
        print("Select the valid option")
        t_movie()
    
    theatre()
```

```python
def theatre():
    print("Choose the theatre in which you want to watch the movie \n 1. INOX \n 2. PVR \n 3. back")
    h = int(input("Enter your theatre name:"))
    if h == 1:
        print("Your theatre is INOX")
    elif h == 2:
        print("Your theatre is PVR")
    elif h == 3:
        t_movie()
    else:
        print("Select the valid option")
        theatre()
    
    screen()
```

```python
def screen():
    print("Choose the screen in which you want to watch the movie \n 1. Screen 1 \n 2. Screen 2 \n 3. Screen 3 \n 4. back")
    s = int(input("Enter your Screen:"))
    if s == 1:
        print("Your screen is Screen 1")
    elif s == 2:
        print("Your screen is Screen 2")
    elif s == 3:
        print("Your screen is Screen 3")
    elif s == 4:
        theatre()
    else:
        print("Select the valid option")
        screen()
    
    timing(s)
```

```python
def timing(s):
    slot1 = {"1" : "10:00-1:00", "2" : "1:15-4:15", "3" : "4:30-7:30", "4" : "7:35-9:35"}
    slot2 = {"1" : "7:45-10:45", "2" : "11:00-2:00", "3" : "2:15-5:15", "4" : "8:20-11:20"}
    slot3 = {"1" : "5:30-8:30", "2" : "8:45-11:45", "3" : "12:00-3:00", "4" : "10:25-1:25"}
    print("Select your timing slot for your movie:\n 1. slot1\n 2. slot2\n 3. slot3\n 4. back")
    if s == 1:
        print("Select your appropriate timing from this slot")
        time = input()
        print("Your selected timing is", slot1[time])
    elif s == 2:
        print("Select your appropriate timing from this slot")
        time = input()
        print("Your selected timing is", slot2[time])
    elif s == 3:
        print("Select your appropriate timing from this slot")
        time = input()
        print("Your selected timing is", slot3[time])
    else:
        print("Select the valid option")
        screen()
        timing(s)
    
    
        
    payment()
    
```

```python
def payment():
    print("Please choose the payment option for your booking \n 1. Card\n 2. UPI\n 3. Debit\n 4. Credit")
    p = int(input("Make your payment:"))
    if p == 1:
        print("You paid through Card")
    elif p == 2:
        print("You paid through UPI")
    elif p == 3:
        print("You paid through Debit")
    elif p == 4:
        print("You paid through Credit")
    else:
        print("Select the valid Option")
        timing(s)
    
        payment()
 
    print("Thank You For Booking your Movie")
    
    
    
    
    
    
    
    
    
```
